```python
import numpy as np
from scipy.special import airy, ai_zeros
import pylab
nmax = 20
a, _, _, _ = ai_zeros(nmax)
qE = -a
m=1
g=9.81
h=1
skalowanie = m*g*(h**(2)/2*g*m**(2))**(1/3)

def prob_qm(n, dq):
    psi, _, _, _ = airy(q-qE[n-1])
    P = psi**2
    return P / (sum(P) * dq)

q, dq = np.linspace(0, 4, 1000, retstep=True)
plt.figure(figsize=(10,6))
pylab.plot(q, prob_qm(1, dq), label='Stan n=1')
pylab.ylim(0,0.8)
pylab.xlim(np.min(q),np.max(q))
plt.grid(linewidth = 1, linestyle = '--')
pylab.xlabel('z')
pylab.ylabel('P(z)')
pylab.legend()
plt.title("Gęstość prawdopodobieństwa w stanie podstawowym")
pylab.savefig("Stan_podstawowy_n1.pdf")
pylab.close()

q, dq = np.linspace(0, 22, 1000, retstep=True)
plt.figure(figsize=(10,6))
pylab.plot(q, prob_qm(20, dq), label='Stan n=20')
pylab.ylim(0,0.25)
pylab.xlim(np.min(q),np.max(q))
plt.grid(linewidth = 1, linestyle = '--')
pylab.xlabel('z')
pylab.ylabel('P(z)')
pylab.legend(loc='upper left')
plt.title("Gęstość prawdopodobieństwa w stanie wzbudzonym n=20")
pylab.savefig("Stan_wzbidzony_n20.pdf")
pylab.close()

q, dq = np.linspace(0, 10, 1000, retstep=True)
plt.figure(figsize=(10,6))
pylab.plot(q, prob_qm(1, dq), label='Stan n=1')
pylab.plot(q, prob_qm(2, dq), label='Stan n=1')
pylab.plot(q, prob_qm(3, dq), label='Stan n=1')
pylab.plot(q, prob_qm(4, dq), label='Stan n=4', ls = '--')
plt.title("Porównanie gęstości prawdopodobieństwa dla 4 stanów stacjonarnych")
pylab.xlim(np.min(q),np.max(q))
plt.grid(linewidth = 1, linestyle = '--')
pylab.xlabel('z')
pylab.ylabel('P(z)')
pylab.legend(loc='upper right')
pylab.savefig("Stan_wzbidzony_n4_podstaowy.pdf")
pylab.close()

#energie przed skalowaniem rowne En = -an
plt.figure(figsize=(10,6))
q, dq = np.linspace(0, 4, 20, retstep=True)
plt.axhline(2.33810741 , 0, 4, label='n=1', color='r')
plt.axhline(4.08794944  , 0, 4, label='n=2', color='y')
plt.axhline(5.52055983  , 0, 4, label='n=3', color='g')
plt.axhline(6.78670809  , 0, 4, label='n=4', color='b')
plt.axhline(7.94413359    , 0, 4, label='n=5', color='k')
plt.axhline(9.02265085 , 0, 4, label='n=6', color='r', ls='--' )
plt.axhline(10.04017434  , 0, 4, label='n=7', color='y', ls='--' )
plt.axhline(11.0085243 , 0, 4, label='n=8', color='g', ls='--' )
plt.axhline(11.93601556 , 0, 4, label='n=9', color='b', ls='--' )
plt.axhline(12.82877675 , 0, 4, label='n=10', color='k', ls='--' )
plt.title("Dozwolone poziomy energetyczne cząstki w polu grawitacyjnym")
pylab.ylabel('Energia')
pylab.legend(loc='center left', bbox_to_anchor=(1, 0.5))
pylab.savefig("Energie_airy.pdf")
pylab.close()

#energie po skalowaniu rowne En = -an*skalowanie
plt.figure(figsize=(10,6))
q, dq = np.linspace(0, 4, 20, retstep=True)
plt.axhline(2.33810741*skalowanie , 0, 4, label='n=1', color='r')
plt.axhline(4.08794944*skalowanie  , 0, 4, label='n=2', color='y')
plt.axhline(5.52055983*skalowanie  , 0, 4, label='n=3', color='g')
plt.axhline(6.78670809*skalowanie  , 0, 4, label='n=4', color='b')
plt.axhline(7.94413359*skalowanie    , 0, 4, label='n=5', color='k')
plt.axhline(9.02265085*skalowanie , 0, 4, label='n=6', color='r', ls='--' )
plt.axhline(10.04017434*skalowanie  , 0, 4, label='n=7', color='y', ls='--' )
plt.axhline(11.0085243*skalowanie , 0, 4, label='n=8', color='g', ls='--' )
plt.axhline(11.93601556*skalowanie , 0, 4, label='n=9', color='b', ls='--' )
plt.axhline(12.82877675*skalowanie , 0, 4, label='n=10', color='k', ls='--' )
plt.title("Poziomy energetyczne cząstki o masie m=1 w polu grawitacyjnym")
pylab.ylabel('Energia')
pylab.legend(loc='center left', bbox_to_anchor=(1, 0.5))
pylab.savefig("Energie_przeskalowane_airy_m1.pdf")
pylab.close()
```
